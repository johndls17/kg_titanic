# Titanic I/O test

Notebook to test Kaggle input-output functionality

```python
import os
import csv
import numpy as np
import pandas as pd
from IPython.display import display

train = pd.read_csv("titanic/train.csv")
test = pd.read_csv("titanic/test.csv")

#verify data contents
display(train.head())
display(test.head())
print(train.shape)
print(test.shape)
```

```python
#prediction = np.empty([len(test),2],dtype=int)
prediction = pd.DataFrame(columns=('PassengerId', 'Survived'))

#a basic loop assigning arbitrary predictions
for x in range(len(test)):
    if test.loc[x,"Age"] > 30:
        prediction.loc[x] = [test.loc[x,"PassengerId"],0]
    else:
        prediction.loc[x] = [test.loc[x,"PassengerId"],1]

#verify prediction characteristics
display(len(prediction))
#display(prediction[:5])
prediction.head()

#convert to DataFrame and publish submission file
#submission = pd.DataFrame({"PassengerId":prediction[:,0],"Survived":prediction[:,1]})
prediction.to_csv("titanic/predictions.csv",index=False)
```
